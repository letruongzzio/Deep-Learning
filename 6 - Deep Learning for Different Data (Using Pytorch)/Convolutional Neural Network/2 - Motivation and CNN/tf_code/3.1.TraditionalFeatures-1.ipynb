```python
import tensorflow as tf
import tensorflow.keras as keras

fashion_mnist = keras.datasets.fashion_mnist
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)
```

```python
X_train = X_train.reshape(-1, 28, 28, 1)
X_test  = X_test.reshape(-1, 28, 28, 1)

X_train = tf.convert_to_tensor(X_train, dtype=tf.float32)
X_test  = tf.convert_to_tensor(X_test, dtype=tf.float32)

print(X_train.shape)
print(X_test.shape)
```

```python
X_train_sobel = tf.image.sobel_edges(X_train)
print(X_train_sobel.shape)

X_test_sobel = tf.image.sobel_edges(X_test)
print(X_test_sobel.shape)
```

```python
X_train_sobel = tf.reshape(X_train_sobel, (-1, 28, 28, 2))
X_test_sobel  = tf.reshape(X_test_sobel, (-1, 28, 28, 2))

print(X_train_sobel.shape)
print(X_test_sobel.shape)
```

```python
# dng v  th, hnh nh
import matplotlib.pyplot as plt 
import numpy as np

# To dang sch 9 phn t ngu nhin t m_train c 60.000 phn t
indices = list(np.random.randint(60000, size=64))

fig = plt.figure(figsize=(9,9))
columns = 8
rows = 8
for i in range(1, columns*rows +1):
    img = tf.reshape(X_train_sobel[indices[i-1],:,:,0], (28, 28))    
    fig.add_subplot(rows, columns, i)
    
    plt.axis('off')
    plt.imshow(img, cmap='gray', vmin=0, vmax=255)
```
